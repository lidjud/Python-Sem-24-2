```python
tuple1 = ("disco",10,1.2 )
tuple1
```

```python
type(tuple1)
```

```python
print(tuple1[0])
print(tuple1[1])
print(tuple1[2])
```

```python
tuple2 = tuple1 + ("hard rock", 10)
tuple2
```

#**Diccionarios**

Un diccionario es una colecci√≥n de pares clave-valor.

lisa-dict.png

## Creaci√≥n de un diccionario

Python usa llaves `{ }` y dos puntos `:` para indicar un diccionario. Puede crear un diccionario vac√≠o y agregar valores m√°s adelante, o bien rellenarlo en el momento de la creaci√≥n. Cada clave o valor est√° separado por dos puntos y el nombre de cada clave se incluye entre comillas como un literal de cadena. Como la clave es un literal de cadena, puede usar el nombre que sea adecuado para describir el valo

```python
# Create the dictionary

Dict = {"key1": 1, "key2": "2", "key3": [3, 3, 3], "key4": (4, 4, 4), ('key5'): 5, (0, 1): 6}
Dict
```

```python
Dict["key1"]
```

```python
Dict[(0, 1)]
```

```python
release_year_dict = {"Thriller": "1982", "Back in Black": "1980", \
                    "The Dark Side of the Moon": "1973", "The Bodyguard": "1992", \
                    "Bat Out of Hell": "1977", "Their Greatest Hits (1971-1975)": "1976", \
                    "Saturday Night Fever": "1977", "Rumours": "1977"}
release_year_dict
```

dict.png

```python
release_year_dict['Thriller']
```

Otro ejemplo...

```python
planet = {
    'name': 'Earth',
    'moons': 1
}
```

Cada clave se comporta igual que una variable: tienen un nombre √∫nico y almacenan un valor.

Puedes leer valores dentro de un diccionario. Los objetos de diccionario tienen un m√©todo get que puede usar para acceder a un valor mediante su clave.

Tambi√©n puede pasar la clave entre corchetes ([ ]).

```python
# planet['name'] is identical to using planet.get('name')
print(planet['name'])
```

```python
wibble = planet.get('wibble') # Returns None
#wibble = planet['wibble'] # Throws KeyError
```

Tambi√©n puede modificar valores dentro de un objeto de diccionario, con el m√©todo `update`

```python
planet.update({'name': 'Makemake'})

# No output: name is now set to Makemake.
```

```python
planet
```

```python
planet['name'] = 'MakeMake'

# No output: name is now set to Makemake.
```

```python
planet
```

```python
planet.update({
    'name': 'Jupiter',
    'moons': 79
})
```

```python
planet
```

```python
planet['name'] = 'Jupiter'
planet['moons'] = 79
```

##Recuperaci√≥n de todas las claves y valores

```python
rainfall = {
    'october': 3.5,
    'november': 4.2,
    'december': 2.1
}
```

```python
for key in rainfall.keys():
    print(f'{key}: {rainfall[key]}cm')
```

```python
if 'december' in rainfall:
    rainfall['december'] = rainfall['december'] + 1
else:
    rainfall['december'] = 1

# Because december exists, the value will be 3.1
```

```python
total_rainfall = 0
for value in rainfall.values():
    total_rainfall = total_rainfall + value

print(f'There was {total_rainfall}cm in the last quarter.')
```
